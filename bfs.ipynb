```python
dict = {
  'S' : [['A',3],['B',6],['C',2]],
  'C' : [['S',2],['E',1]],
  'A' : [['D',3], ['S',3]],
  'B' : [['D',4],['G',9],['E',2],['S',6]],
  'D' : [['A',3],['B',4],['F',5]],
  'E' : [['F',6],['H',5],['C',1],['B',2]],
  'F' : [['G',5],['E',6],['D',5]],
  'G' : [['F',5],['B',9],['H',8]],
  'H' : [['E',5],['G',8]]
}
#print(len(dict))
#print(dict.keys())
#print(list(dict.keys())[0])
#print(list(dict.values())[1][1])
# k = list(dict.keys())[0]
# print(k)
# f = list(dict.values())[0]
# print(f)
def Sort_List_Of_List(put_Val):
    #it tells us the number of elements in the outer list of the root node
    lengthOfBasic = len(put_Val)

    #print(lengthOfBasic)

    for i in range(0, lengthOfBasic):
    
        for j in range(0, lengthOfBasic-i-1):
           
            if (put_Val[j][1] > put_Val[j + 1][1]):
                # print(put_Val[j+1][1])
                # print(put_Val[j][1])
                # print(put_Val[j])
                # print(put_Val[j+1])
                # print(put_Val[j-1])
                # print(put_Val[j-1][1])
                # print(put_Val[j][0])
                

               temp = put_Val[j]
               put_Val[j]= put_Val[j + 1]
               put_Val[j + 1]=temp
             
    return put_Val
def bfs(dict, start_point):
    valll = []
    final_list = [] 
    final_list.append(start_point)
    for i in range(len(dict)):
         #print(list(dict.keys())[i])
         #print(list(dict.values())[i])
          # In this case my start point is S
        if( start_point == list(dict.keys())[i]):
          # it stores the list of list of starting point 
            eleeee = len(list(dict.values())[i])
            # print(list(dict.keys())[i])
            # print(list(dict.values())[i])

            #it stores the list of list values in A
            A = list(dict.values())[i]
            #print(A)
            #Sorts the list of list distance value in ascending order
            A = Sort_List_Of_List(A)
            #print(A)


            for j in range(eleeee):
                #it sorts the the nodes value according to distance (i.e C A B)
                keeeey = (A[j][0])
                #it puts C, A,B one by one in the valll list
                valll.append(keeeey)
                #print(keeeey)
                #print(valll)
            try: 
                #it removes left most element from the valll list i.e C (FIFO)             
                CCC = valll.pop(0)
                #print(CCC)
                final_list.append(CCC)
                # it adds the element which is removed from valll list in final list
                
                #print(final_list)
                # after c is removed from valll it is left with A and B
                # print(valll)
            except:
                pass
    #print(dict[CCC])  
    while True:
        for nnnn in dict[CCC]:
        #      print(nnnn)
        # break    
            if nnnn[0] not in valll and nnnn[0] not in final_list:
                valll.append(nnnn[0])
                # print(nnnn[0])
        if len(valll) == 0:
            break         
        CCC = valll.pop(0)
        final_list.append(CCC)    
    print(final_list)
bfs(dict,"S")
```
